# I will analyze Divvy Data in Rush Hour

```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# save file path to variable
q_one = "Divvy_Trips_2017_Q1.csv"

# read with pandas
q_one = pd.read_csv(q_one)
q_one.head()
```

```python
# save file path to variable
q_two = "Divvy_Trips_2017_Q2.csv"

# read with pandas
q_two = pd.read_csv(q_two)
q_two.head()
```

```python
# save file path to variable
q_three = "Divvy_Trips_2017_Q3.csv"

# read with pandas
q_three = pd.read_csv(q_three)
q_three.head()
```

```python
# save file path to variable
q_four = "Divvy_Trips_2017_Q4.csv"

# read with pandas
q_four = pd.read_csv(q_four)
q_four.head()
```

# Create Divvy Dataframe from trip divvy data

```python
# join tables
# pd.concat(objs, axis=0, join='outer', join_axes=None, ignore_index=False,
#           keys=None, levels=None, names=None, verify_integrity=False,
#           copy=True)
divvy_data = pd.concat([q_one, q_two, q_three, q_four])
divvy_data
```

```python
# save file path to variable
stations = "Divvy_Stations_2017_Q3Q4.csv"

# read with pandas
stations = pd.read_csv(stations)
stations.head()
```

```python
# rename columns
stationsfrom = stations.rename(columns = {"id":"from_station_id"})
stationsfrom.head()
```

# Merge location data with divvy data

```python
# combine tables
div_data = pd.merge(divvy_data, stationsfrom, on ='from_station_id')
div_data
```

```python
# rename columns
div_data = div_data.rename(columns = 
                          {"online_date":"from_ondate"})
```

```python
div_data
```

```python
# rename columns
stationsto = stations.rename(columns = {"id":"to_station_id"})
stationsto.head()
```

```python
# combine tables
divvy_df = pd.merge(div_data, stationsto, on ='to_station_id')
divvy_df.shape
```

```python
# rename columns
divvy_df = divvy_df.rename(columns = {"online_date":"to_ondate"})
divvy_df.head()
```

```python
# change the data format so it is actually datetime
divvy_df["start_time"] = pd.to_datetime(divvy_df["start_time"])
divvy_df["end_time"] = pd.to_datetime(divvy_df["end_time"])
divvy_df["start_month"] = divvy_df["start_time"].map(lambda x: x.month)
divvy_df["end_month"] = divvy_df["end_time"].map(lambda x: x.month)
divvy_df.head()
```

```python
# # # df = df.drop('column_name', 1)
# divvy_df = divvy_df.drop(columns=['month'])
divvy_df["start_date"] = divvy_df['start_time']
divvy_df.head()
```

```python
#make dates format
divvy_df['start_date'] = divvy_df['start_date'].dt.date
```

```python
#drop blank columns
divvy_df = divvy_df.drop(columns=['Unnamed: 7_x'])
divvy_df = divvy_df.drop(columns=['Unnamed: 7_y'])
list(divvy_df)
```

```python
# format dates
divvy_df['end_date'] = divvy_df['end_time'].dt.date
```

# Bring in data breaking out holidays, weekends, and workdays so we only include workdays in our analysis

```python
# save file path to variable
holidays = "2017_Holidays.csv"

# read with pandas
holidays = pd.read_csv(holidays)
holidays.head()
```

```python
# match date name
holidays = holidays.rename(columns = {"Date":"start_date"})
holidays.head()
```

```python
# change format to date
holidays["start_date"] = pd.to_datetime(holidays["start_date"])
holidays['start_date'] = holidays['start_date'].dt.date
```

```python
# combine tables
combined_df = pd.merge(divvy_df, holidays, on ='start_date')
combined_df.head()
```

```python
combined_df.head()
```

```python
# remove holidays and weekends
weekdays_df = combined_df[combined_df["Include/Not_Include"] =='Weekday']
weekdays_df.head()
```

```python
# format time
weekdays_df['begin_time'] = weekdays_df['start_time'].dt.time
weekdays_df['stop_time'] = weekdays_df['end_time'].dt.time
```

```python
weekdays_df.head()
```

# Create hour bins so we can figure out rush hour times

```python
# add the hour of the day for binning later
weekdays_df["hour"] = weekdays_df["begin_time"].apply(lambda x: x.hour)
```

```python
weekdays_df.head()
```

```python
# Bins are Midnight to 6am, 6am to 10am, 10am to 4pm, 4pm to 8pm, 8pm to 12pm
bins = [0, 5, 10, 15, 20, 23]

# Create the names for the four bins
group_names = ['early', 'morning_rh', 'midday', 'afterwork_rh', 'evening']
```

```python
# Cut hour and place the hour into bins
pd.cut(weekdays_df["hour"], bins, labels=group_names)
```

```python
weekdays_df["Time_of_Day"] = pd.cut(
    weekdays_df["hour"], bins, labels=group_names)
```

```python
weekdays_df.head()
```

# Create the Rush Hour Dataframe

```python
# Create Rush Hour Dataframe
mrh_df = weekdays_df[weekdays_df["Time_of_Day"] =='morning_rh']
erh_df = weekdays_df[weekdays_df["Time_of_Day"] =='afterwork_rh']

# concat dfs together
rh_df = pd.concat([mrh_df, erh_df])
```

```python
#reset index
rh_df = rh_df.reset_index(drop=True)
```

```python
#drop blank columns
# rh_df = rh_df.drop(columns=['index'])
# rh_df.head(10)
```

# Create age bins so we can isolate working ages during rush hour

```python
# Bins are 0 to 1954, 1955 to 1996, 1997 to present
of_age = [0, 1954, 1996, 2018]

# Create the names for the four bins
group_names2 = ['retired', 'working_age', 'children']
```

```python
# Cut hour and place the years into bins
pd.cut(rh_df["birthyear"], of_age, labels=group_names2)
```

```python
rh_df["Of Working Age?"] = pd.cut(
    rh_df["birthyear"], of_age, labels=group_names2)
```

```python
rh_df.head()
```

# Find the average trip duration per hour during rush hour

```python
# average trip duration by hour
group_hour = rh_df.groupby(by="hour")
group_hour
# avgs
hour_avg_ride = (group_hour["tripduration"].mean())/60
hour_avg_ride
# scatter of hour of day and tripduration
# plt.scatter(rh_df["hour"], rh_df["tripduration"], color = "lightcoral", edgecolors="black", alpha = 0.5, linewidth = 1)
#bar chart of duration avgs
hour_avg_ride.plot.bar()
plt.title("Average Trip Duration Per Hour During Rush Hour")
plt.ylabel("Trip Duration (minutes)")
plt.savefig("outputs/avgtripdurationbarchart.png", bbox_inches='tight')

#Add range and color scheme
# plt.grid(color='white')
# plt.ylim(15, 55)
plt.style.use('seaborn')
```

```python
# average trip duration by group
group_age = rh_df.groupby(by="Of Working Age?")
group_age
# avgs
age_avg_ride = (group_age["tripduration"].mean())/60
age_avg_ride
```

# Create chart showing average trip duration during rush hour by age group

```python
#bar chart of duration avgs
age_avg_ride.plot.bar()
plt.title("Average Trip Duration in Minutes During Rush Hour")
plt.ylabel("Minutes")
plt.savefig("outputs/tripgroupsbarchart.png", bbox_inches='tight')
```

```python
age_count_trips = group_age["trip_id"].count()
age_count_trips
```

# Find percent of total trips taken by age group during rush hour

```python
# pie chart of rush hour rides
plt.title("Percent of Total Trips During Rush Hour")
labels = ["Retired", "Working Age", "Children"]
colors = ['orange', 'royalblue', 'turquoise']
explode = [.7, 0, .7]
# plt.ylabel("Minutes")
#chart effects
plt.pie(age_count_trips, explode = explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.axis("equal")
plt.savefig("outputs/percenttrips.png", bbox_inches='tight')
```

# Create Working Age only Rush Hour Dataframe

```python
# Create Rush Hour Dataframe
rush_hour = rh_df[rh_df["Of Working Age?"] =='working_age']
rush_hour.head()
```

```python
# group by station
Fstation_group = rush_hour.groupby(by="from_station_name")
```

```python
# determine avg ride time in minutes from station
avg_ride = (Fstation_group["tripduration"].mean())/60
avg_ride.head()
```

# Find the middle 10 avg ride times between stations of all ride trip groups

```python
# group by station
Tstation_group = rush_hour.groupby(["from_station_name", "to_station_name"])

# determine middle 10 avg ride times in minutes between stations
avg_ride_mid = (Tstation_group["tripduration"].mean())/60
avg_ride_mid = avg_ride_mid.sort_values(ascending = False)
# avg_rideT = avg_rideT.mode()
avg_ride_mid.iloc[36400:36411]
mid_avg_ride = avg_ride_mid.iloc[36400:36411]
mid_avg_ride.to_csv("outputs/mid_avg_ride.csv", header=True, sep='\t', encoding='utf-8')
mid_avg_ride
```

# Find the top 10 avg ride times between stations

```python
# determine top 10 avg ride times in minutes to station
avg_rideT = (Tstation_group["tripduration"].mean())/60
avg_rideT = avg_rideT.sort_values(ascending = False)
avg_rideT = avg_rideT.head(10)
avg_rideT.to_csv("outputs/top_10_avg_rides.csv", header=True, sep='\t', encoding='utf-8')
avg_rideT
```

```python
# group by station
station_bygroup = rush_hour.groupby(["from_station_name", "to_station_name"])
```

```python
#count trips between each station
grouped_stations = station_bygroup['from_station_id'].value_counts()
grouped_stations.head()
```

```python
#get top ten trips to and from and count of trips
high_vol_trips = grouped_stations.sort_values(ascending = False)
high_vol_trips = high_vol_trips.head(10)
high_vol_trips.to_csv("outputs/high_vol_trips.csv", header=True, sep='\t', encoding='utf-8')
high_vol_trips
```

# Plot the top 10 trip routes taken by total count
