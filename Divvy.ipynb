```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# save file path to variable
q_one = "Divvy_Trips_2017_Q1.csv"

# read with pandas
q_one = pd.read_csv(q_one)
q_one.head()
```

```python
# save file path to variable
q_two = "Divvy_Trips_2017_Q2.csv"

# read with pandas
q_two = pd.read_csv(q_two)
q_two.head()
```

```python
# save file path to variable
q_three = "Divvy_Trips_2017_Q3.csv"

# read with pandas
q_three = pd.read_csv(q_three)
q_three.head()
```

```python
# save file path to variable
q_four = "Divvy_Trips_2017_Q4.csv"

# read with pandas
q_four = pd.read_csv(q_four)
q_four.head()
```

```python
# join tables
# pd.concat(objs, axis=0, join='outer', join_axes=None, ignore_index=False,
#           keys=None, levels=None, names=None, verify_integrity=False,
#           copy=True)
divvy_data = pd.concat([q_one, q_two, q_three, q_four])
divvy_data
```

```python
# save file path to variable
stations = "Divvy_Stations_2017_Q3Q4.csv"

# read with pandas
stations = pd.read_csv(stations)
stations.head()
```

```python
# rename columns
stationsfrom = stations.rename(columns = {"id":"from_station_id"})
stationsfrom.head()
```

```python
# combine tables
div_data = pd.merge(divvy_data, stationsfrom, on ='from_station_id')
div_data = div_data.head()
```

```python
div_data
```

```python
# rename columns
div_data = div_data.rename(columns = 
                          {"online_date":"from_ondate"})
```

```python
# combine tables
divvy_df = pd.merge(div_data, stationsto, on ='to_station_id')
divvy_df.head()
```
