# A Series of Graphs

```python
#Import Dependencies
import pandas as pd 
import folium
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
csv_file = os.path.join("outputs/divvy_data_2017.csv")
master_data_2017 = pd.read_csv(csv_file)
gender_data = master_data_2017.groupby(["gender"]).count()
```

```python
master_data_2017["start_time"] = pd.to_datetime(master_data_2017["start_time"])
master_data_2017["end_time"] = pd.to_datetime(master_data_2017["end_time"])
master_data_2017["hour"] = master_data_2017["start_time"].map(lambda x: x.hour)
master_data_2017
```

# Female vs. Male Riders

```python
sns.barplot(gender_data['usertype'],gender_data.index, palette="Blues_d")
plt.title("2017 Divvy Usage by Gender")
plt.savefig('outputs/usagebygender.png')
```

```python
avg_trip = master_data_2017['tripduration'].mean()
avg_trip
```

# Times of the Day

```python
hours = master_data_2017.groupby('hour').count()
hours.head()
```

```python
sns.barplot(hours.index, hours['trip_id'])
plt.grid()
plt.xticks(rotation = 'vertical')
plt.title('Ridership per hour in 2017')
plt.ylabel('Num of rides')
plt.savefig('outputs/ridershiphour.png')
```

```python
csv_station = os.path.join("Data/Divvy_Stations_2017_Q3Q4.csv")
station_df = pd.read_csv(csv_station)
```

# Popular Stations

```python
pop_locations = master_data_2017.groupby(['from_station_id']).count()
pop_locations = pop_locations.sort_values(by='trip_id',ascending =False).reset_index()
pop_locations = pop_locations[0:10].rename(columns = {"from_station_id":"id","bikeid":"number_of_rides"})
pop_locations = pop_locations.drop(columns = ['Unnamed: 0', 'trip_id', 'start_time', 'end_time',
       'tripduration', 'from_station_name', 'to_station_id',
       'to_station_name', 'usertype', 'gender', 'birthyear', 'month', 'hour'])
pop_locations = pop_locations.merge(station_df, on="id")
```

```python
plt.pie(pop_locations['number_of_rides'], labels=pop_locations['name'])
plt.axis("equal")
plt.title("Top 10 Most Popular Stations Overall")
plt.tight_layout()
plt.savefig("piegraph.png")
None
```
